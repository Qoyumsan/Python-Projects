```python
product_line_total = df.groupby("Product_line")["Total"].sum()
plt.figure(figsize=(10,9))
bars = plt.bar(product_line_total.index, product_line_total.values, color ="skyblue")
plt.xlabel("Product_line")
plt.ylabel("Total")
plt.title("Total Amount Per Product Line")
plt.xticks(rotation = 45, ha= "right")
plt.show()
```

```python
product_line_total = df.groupby("City")["Total"].sum()
plt.figure(figsize=(10,9))
bars = plt.bar(product_line_total.index, product_line_total.values, color ="blue")
plt.xlabel("City")
plt.ylabel("Total")
plt.title("Total Amount Per City")
plt.xticks(rotation = 45, ha= "right")
plt.show()
```

```python
product_line_total = df.groupby("medium_of_payment")["Total"].sum()
plt.figure(figsize=(9,6))
bars = plt.bar(product_line_total.index, product_line_total.values, color ="skyblue")
plt.xlabel("Medium_of_payment")
plt.ylabel("Total")
plt.title("Total Amount by medium_of_payment")
plt.xticks(rotation = 45, ha= "right")
plt.show()
```
